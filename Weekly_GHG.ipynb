```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_tracker = pd.read_excel("https://github.com/NicolasWoloszko/OECD-Weekly-Tracker/raw/main/Data/weekly_tracker.xlsx")
```

```python
df_tracker = df_tracker[['region', 'date', 'Tracker (yoy)']]
df_tracker = df_tracker.rename(columns = {"region": "Country", 
                                          "date": "Week", 
                                          "Tracker (yoy)": "Change"})
df_tracker["Week"] = df_tracker["Week"].astype(str)
df_tracker
```

```python
df_tracker["Country"].unique()
```

```python
countries = ['United Kingdom', 'United States', 'France', 'Germany']
```

```python
weeks = ['2021-11-07', '2021-11-14']
```

```python
df_tracker = df_tracker[df_tracker["Country"].apply(lambda x: x in countries)]
df_tracker = df_tracker[df_tracker["Week"].apply(lambda x: x in weeks)]

df_tracker = df_tracker.reset_index(drop = True)

df_tracker
```

```python
df_tracker = df_tracker.reset_index(drop = True)
df_tracker
```

```python
years = [2017, 2018, 2019]
```

```python
df_ghg = pd.read_csv("AIR_GHG_25112021005547428.csv")
df_ghg = df_ghg[df_ghg["Variable"] == 'Total  emissions excluding LULUCF']
df_ghg = df_ghg[df_ghg["Unit"] == 'Tonnes of CO2 equivalent']
df_ghg = df_ghg[df_ghg["Pollutant"] == 'Greenhouse gases']
df_ghg = df_ghg[df_ghg["Year"].apply(lambda x: x in years)]
df_ghg = df_ghg[df_ghg["Country"].apply(lambda x: x in countries)]

df_ghg = df_ghg[["Country", "Year", "Value"]]

df_ghg = df_ghg.reset_index(drop = True)
df_ghg
```

```python
df_average = df_ghg[["Country", "Value"]].groupby(by = "Country").agg("mean")
df_average
```

```python
for i in range(len(df_tracker)):
    historic_emission = (df_average.loc[df_tracker.loc[i, "Country"], "Value"]/52)
    weekly_emission = historic_emission*(1.0 + df_tracker.loc[i, "Change"]/100)
    df_tracker.loc[i, "weekly_emission"] = weekly_emission
    print(weekly_emission)
```
